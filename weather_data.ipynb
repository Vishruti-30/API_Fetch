# üîπ Step 2: Set your API key and base URL

# üåç Step 3: Define List of Cities
## You can edit this list to include any cities you want to monitor.

```python
cities = ["Toronto", "New York", "London", "Paris", "Tokyo"]
```

# üîÑ Step 4: Fetch Weather Data for Each City
## For each city, we'll make a request to the API and extract key information like temperature, description, humidity, and wind speed.


```python
all_weather_data = {}

for city in cities:
    params = {
        'q': city,
        'appid': API_KEY,
        'units': 'metric'
    }

    response = requests.get(BASE_URL, params=params)
    
    if response.status_code == 200:
        data = response.json()
        all_weather_data[city] = {
            "temperature": data['main']['temp'],
            "description": data['weather'][0]['description'],
            "humidity": data['main']['humidity'],
            "wind_speed": data['wind']['speed']
        }
    else:
        all_weather_data[city] = {
            "error": f"Could not fetch data (status code {response.status_code})"
        }
```

# üíæ Step 5: Save Data to JSON File
## All weather data will be saved to `weather_data.json` in the current directory.

```python
with open("weather_data.json", "w") as f:
    json.dump(all_weather_data, f, indent=4)

print("‚úÖ Weather data saved to weather_data.json")
```
